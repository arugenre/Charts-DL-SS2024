```python
# templates_line_charts = ["Create a {chart_type} chart using ChartJS. The chart should be titled '{title}'. It will display data points for specific dates: {dates}. The corresponding values for these dates are {values} respectively.",
#              "Craft a visual representation with ChartJS, ensuring the title reads '{title}'. This graph will showcase the stock's highest values on the specified dates: {date_value_key_pairs}.",
#              "Craft a {chart_type} graph with ChartJS, naming it '{title}'. This graph will showcase the data for certain key dates: {dates}. The values associated with these dates are as follows: {values}.",
#              "Craft a {chart_type} graph with ChartJS, naming it '{title}'. This graph will present figures for certain chronological markers: {dates}. The values linked to these dates are {values} respectively.",
#              "Create a visual representation with ChartJS, titled '{title}'. This chart will illustrate the data for particular moments: {date_value_key_pairs}.",
#              "Craft a {chart_type} graph with ChartJS, titling it '{title}'. This visualization will present data markers for particular times: {dates}. The associated figures for these dates are {values}, respectively.",
#              "Create a {chart_type} graph with ChartJS, titled '{title}'. This graph will showcase markers for certain time periods: {dates}. The values associated with these dates are as follows: {values}.",
#              "Create a {chart_type} graph with ChartJS, titled '{title}'. This graph will represent data markers for certain chronological instances: {dates}. The associated figures for these dates are {values}, respectively.",
#              "Construct a {chart_type} chart utilizing ChartJS titled '{title}' with the following data: For the dates of {date_value_key_pairs}."
#             ]

# templates_pie_charts = [
#             "Create a {chart_type} chart using ChartJS. The chart should be titled '{title}'. It will display data points for specific countries: {xvalues}. The corresponding happiness index for these countries are {yvalues} respectively.",
#             "Craft a visual representation with ChartJS, ensuring the title reads '{title}'. This graph will showcase the happiness index for the specified countries: {xvalue_yvalue_key_pairs}.",
#             "Craft a {chart_type} graph with ChartJS, naming it '{title}'. This graph will showcase the data for certain countries: {xvalues}. The happiness index associated with these countries are as follows: {yvalues}.",
#             "Craft a {chart_type} graph with ChartJS, naming it '{title}'. This graph will present figures for certain geographical markers: {xvalues}. The happiness index linked to these countries are {yvalues} respectively.",
#             "Create a visual representation with ChartJS, titled '{title}'. This chart will illustrate the data for particular countries: {xvalue_yvalue_key_pairs}.",
#             "Craft a {chart_type} graph with ChartJS, titling it '{title}'. This visualization will present data markers for particular countries: {xvalues}. The associated happiness index for these countries are {yvalues}, respectively.",
#             "Create a {chart_type} graph with ChartJS, titled '{title}'. This graph will showcase markers for certain countries: {xvalues}. The happiness index associated with these countries are as follows: {yvalues}.",
#             "Create a {chart_type} graph with ChartJS, titled '{title}'. This graph will represent data markers for certain geographical instances: {xvalues}. The associated happiness index for these countries are {yvalues}, respectively.",
#             "Construct a {chart_type} chart utilizing ChartJS titled '{title}' with the following data: For the countries of {xvalue_yvalue_key_pairs}."
# ]

# templates_bar_charts = [
#     "Create a {chart_type} chart using ChartJS. The chart should be titled '{title}'. It will display data points for specific countries: {xvalues}. The corresponding burnt area values for these countries are {yvalues}, respectively.",
#     "Craft a visual representation with ChartJS, ensuring the title reads '{title}'. This graph will showcase the burnt area in hectares for the specified countries: {xvalue_yvalue_key_pairs}.",
#     "Craft a {chart_type} graph with ChartJS, naming it '{title}'. This graph will showcase the data for certain countries: {xvalues}. The burnt area values associated with these countries are as follows: {yvalues}.",
#     "Craft a {chart_type} graph with ChartJS, naming it '{title}'. This graph will present figures for certain countries: {xvalues}. The values linked to the burnt area in these countries are {yvalues}, respectively.",
#     "Create a visual representation with ChartJS, titled '{title}'. This chart will illustrate the burnt area data for particular countries: {xvalue_yvalue_key_pairs}.",
#     "Craft a {chart_type} graph with ChartJS, titling it '{title}'. This visualization will present data markers for specific countries: {xvalues}. The associated burnt area values for these countries are {yvalues}, respectively.",
#     "Create a {chart_type} graph with ChartJS, titled '{title}'. This graph will showcase markers for certain countries: {xvalues}. The values associated with the burnt area for these countries are as follows: {yvalues}.",
#     "Create a {chart_type} graph with ChartJS, titled '{title}'. This graph will represent data markers for specific countries: {xvalues}. The burnt area values associated with these countries are {yvalues}, respectively.",
#     "Construct a {chart_type} chart utilizing ChartJS titled '{title}' with the following data: For the countries of {xvalue_yvalue_key_pairs}."
# ] 

# Templates for line chart data(Which is based on stocks data and their price throughout a time period)
templates_line_charts = [
    "Create a {chart_type} chart using ChartJS. The chart should be titled '{title}'. It will display data points for specific dates: {dates}. The corresponding values for these dates are {values} respectively.",
    "Craft a visual representation with ChartJS, ensuring the title reads '{title}'. This graph will showcase the stock's highest values on the specified dates: {date_value_key_pairs}.",
    "Craft a {chart_type} graph utilizing ChartJS, naming it '{title}'. This graph will showcase the data for certain key dates: {dates}. The values associated with these dates are as follows: {values}.",
    "Craft a {chart_type} graph utilizing ChartJS, naming it '{title}'. This graph will present figures for certain chronological markers: {dates}. The values linked to these dates are {values} respectively.",
    "Create a visual representation with ChartJS, titled '{title}'. This chart will illustrate the data for particular moments: {date_value_key_pairs}.",
    "Craft a {chart_type} graph utilizing ChartJS, titling it '{title}'. This visualization will present data markers for particular times: {dates}. The associated figures for these dates are {values}, respectively.",
    "Create a {chart_type} graph utilizing ChartJS, titled '{title}'. This graph will showcase markers for certain time periods: {dates}. The values associated with these dates are as follows: {values}.",
    "Create a {chart_type} graph utilizing ChartJS, titled '{title}'. This graph will represent data markers for certain chronological instances: {dates}. The associated figures for these dates are {values}, respectively.",
    "Construct a {chart_type} chart utilizing ChartJS titled '{title}' with the following data: For the dates of {date_value_key_pairs}.",
    "Generate a {chart_type} chart using ChartJS and label it '{title}'. This chart will highlight data points for key dates: {dates}. The values for these dates are {values}.",
    "Design a {chart_type} chart through ChartJS, titled '{title}'. This chart will display data points for important dates: {dates}. The respective values are {values}.",
    "Create a {chart_type} chart using ChartJS, with the title '{title}'. The chart will show data for specific dates: {dates}, with the corresponding values being {values}.",
    "Develop a {chart_type} chart through ChartJS, named '{title}'. This visualization will depict data for specific dates: {dates}, with their respective values: {values}.",
    "Construct a {chart_type} chart using ChartJS titled '{title}'. This chart will illustrate data points for certain dates: {dates}. The associated values are {values}.",
    "Generate a visual with ChartJS titled '{title}'. This {chart_type} chart will display data points for the dates: {dates}, with corresponding values of {values}.",
    "Design a {chart_type} chart with ChartJS, with the title '{title}'. This chart will show data points for specific dates: {dates}. The respective values are {values}.",
    "Create a {chart_type} chart using ChartJS, titled '{title}'. This visualization will showcase data points for the following dates: {dates}, with their associated values being {values}.",
    "Develop a {chart_type} chart implementing ChartJS, titled '{title}'. The chart will present data for specific dates: {dates}, along with the corresponding values: {values}.",
    "Build a {chart_type} chart implementing ChartJS and name it '{title}'. This chart will highlight data points for the dates: {dates}, with the respective values being {values}.",
    "Construct a {chart_type} chart using ChartJS, named '{title}'. This chart will show data for the dates: {dates}, with the corresponding values being {values}.",
    "Assemble a {chart_type} chart implementing ChartJS and title it '{title}'. This visualization will highlight data points for specific dates: {dates}, with their respective values being {values}.",
    "Develop a {chart_type} chart implementing ChartJS, named '{title}'. The chart will display data points for significant dates: {dates}, with their corresponding values: {values}.",
    "Create a {chart_type} chart leveraging ChartJS titled '{title}' with this data(date value pairs): Dates and values are {date_value_key_pairs}.",
    "Generate a {chart_type} chart through ChartJS titled '{title}' using these date value pairs: {date_value_key_pairs}.",
    "Design a {chart_type} chart in ChartJS notation titled '{title}' showcasing the following dates and their values: {date_value_key_pairs}.",

]

# Templates for pie chart data(Which is based on happiness index data and their value between several countries)
templates_pie_charts = [
    "Create a {chart_type} chart utilizing ChartJS. The chart should be titled '{title}'. It will display data points for specific countries: {xvalues}. The corresponding happiness index for these countries are {yvalues} respectively.",
    "Craft a visual representation utilizing ChartJS, ensuring the title reads '{title}'. This graph will showcase the happiness index for the specified countries: {xvalue_yvalue_key_pairs}.",
    "Craft a {chart_type} graph utilizing ChartJS, naming it '{title}'. This graph will showcase the data for certain countries: {xvalues}. The happiness index associated with these countries are as follows: {yvalues}.",
    "Craft a {chart_type} graph utilizing ChartJS, naming it '{title}'. This graph will present figures for certain geographical markers: {xvalues}. The happiness index linked to these countries are {yvalues} respectively.",
    "Create a visual representation utilizing ChartJS, titled '{title}'. This chart will illustrate the data for particular countries: {xvalue_yvalue_key_pairs}.",
    "Craft a {chart_type} graph utilizing ChartJS, titling it '{title}'. This visualization will present data markers for particular countries: {xvalues}. The associated happiness index for these countries are {yvalues}, respectively.",
    "Create a {chart_type} graph utilizing ChartJS, titled '{title}'. This graph will represent data markers for certain geographical instances: {xvalues}. The associated happiness index for these countries are {yvalues}, respectively.",
    "Construct a {chart_type} chart utilizing ChartJS titled '{title}' with the following data: For the countries of {xvalue_yvalue_key_pairs}.",
    "Generate a {chart_type} chart utilizing ChartJS and label it '{title}'. This chart will highlight data points for key countries: {xvalues}. The happiness index for these countries are {yvalues}.",
    "Develop a {chart_type} chart utilizing ChartJS, named '{title}'. This visualization will depict data for specific countries: {xvalues}, with their respective happiness index values: {yvalues}.",
    "Construct a {chart_type} chart utilizing ChartJS titled '{title}'. This chart will illustrate data points for certain countries: {xvalues}. The associated happiness index values are {yvalues}.",
    "Design a {chart_type} chart utilizing ChartJS, with the title '{title}'. This chart will show data points for specific countries: {xvalues}. The respective happiness index values are {yvalues}.",
    "Assemble a {chart_type} chart executing ChartJS titled '{title}' to present data as follows: {xvalue_yvalue_key_pairs}.",
    "Develop a {chart_type} chart utilizing ChartJS, titled '{title}'. The chart will present data for specific countries: {xvalues}, along with the corresponding happiness index values: {yvalues}.",
    "Build a {chart_type} chart utilizing ChartJS and name it '{title}'. This chart will highlight data points for the countries: {xvalues}, with the respective happiness index values being {yvalues}.",
    "Construct a {chart_type} chart utilizing ChartJS, named '{title}'. This chart will show data for the countries: {xvalues}, with the corresponding happiness index values being {yvalues}.",
    "Assemble a {chart_type} chart utilizing ChartJS and title it '{title}'. This visualization will highlight data points for specific countries: {xvalues}, with their respective happiness index values being {yvalues}.",
    "Develop a {chart_type} chart utilizing ChartJS, named '{title}'. The chart will display data points for significant countries: {xvalues}, with their corresponding happiness index values: {yvalues}.",
    "Formulate a {chart_type} chart utilizing ChartJS, titled '{title}'. This graph will present data points for given countries: {xvalues}, with their respective happiness index values being {yvalues}.",
    "Assemble a {chart_type} chart executing ChartJS titled '{title}' to present data as follows: {xvalue_yvalue_key_pairs}.",
    "Generate a {chart_type} chart utilizing ChartJS, titled '{title}'. This chart will illustrate data points for key countries: {xvalues}, and the happiness index values for these countries are {yvalues}.",
    "Design a {chart_type} graph utilizing ChartJS, named '{title}'. This chart will show data points for particular countries: {xvalues}, and their respective happiness index values are {yvalues}.",
    "Assemble a {chart_type} chart executing ChartJS titled '{title}' to present data as follows: {xvalue_yvalue_key_pairs}.",
    "Assemble a {chart_type} chart utilizing ChartJS, named '{title}'. This visualization will display data points for the following countries: {xvalues}, with the associated happiness index values being {yvalues}.",
    "Create a {chart_type} chart utilizing ChartJS, titled '{title}'. This chart will show data points for certain countries: {xvalues}, and the corresponding happiness index values are {yvalues}.",
    "Assemble a {chart_type} chart executing ChartJS titled '{title}' to present data as follows: {xvalue_yvalue_key_pairs}."
]

# Templates for bar chart data(Which is based on are burned in widlfires between several countries)
templates_bar_charts = [
    "Create a {chart_type} chart utilizing ChartJS. The chart should be titled '{title}'. It will display data points for specific countries: {xvalues}. The corresponding burnt area values for these countries are {yvalues}, respectively.",
    "Craft a visual representation utilizing ChartJS, ensuring the title reads '{title}'. This graph will showcase the burnt area in hectares for the specified countries: {xvalue_yvalue_key_pairs}.",
    "Craft a {chart_type} graph utilizing ChartJS, naming it '{title}'. This graph will showcase the data for certain countries: {xvalues}. The burnt area values associated with these countries are as follows: {yvalues}.",
    "Craft a {chart_type} graph utilizing ChartJS, naming it '{title}'. This graph will present figures for certain countries: {xvalues}. The values linked to the burnt area in these countries are {yvalues}, respectively.",
    "Create a visual representation utilizing ChartJS, titled '{title}'. This chart will illustrate the burnt area data for particular countries: {xvalue_yvalue_key_pairs}.",
    "Craft a {chart_type} graph utilizing ChartJS, titling it '{title}'. This visualization will present data markers for specific countries: {xvalues}. The associated burnt area values for these countries are {yvalues}, respectively.",
    "Create a {chart_type} graph utilizing ChartJS, titled '{title}'. This graph will showcase markers for certain countries: {xvalues}. The values associated with the burnt area for these countries are as follows: {yvalues}.",
    "Create a {chart_type} graph utilizing ChartJS, titled '{title}'. This graph will represent data markers for specific countries: {xvalues}. The burnt area values associated with these countries are {yvalues}, respectively.",
    "Construct a {chart_type} chart utilizing ChartJS titled '{title}' with the following data: For the countries of {xvalue_yvalue_key_pairs}.",
    "Generate a {chart_type} chart utilizing ChartJS and label it '{title}'. This chart will highlight data points for key countries: {xvalues}. The burnt area values for these countries are {yvalues}.",
    "Design a {chart_type} chart utilizing ChartJS, titled '{title}'. This chart will display data points for important countries: {xvalues}. The respective burnt area values are {yvalues}.",
    "Create a {chart_type} chart utilizing ChartJS, with the title '{title}'. The chart will show data for specific countries: {xvalues}, with the corresponding burnt area being {yvalues}.",
    "Develop a {chart_type} chart utilizing ChartJS, named '{title}'. This visualization will depict data for specific countries: {xvalues}, with their respective burnt area values: {yvalues}.",
    "Construct a {chart_type} chart utilizing ChartJS titled '{title}'. This chart will illustrate data points for certain countries: {xvalues}. The associated burnt area values are {yvalues}.",
    "Generate a visual utilizing ChartJS titled '{title}'. This {chart_type} chart will display data points for the countries: {xvalues}, with corresponding burnt area values of {yvalues}.",
    "Design a {chart_type} chart utilizing ChartJS, with the title '{title}'. This chart will show data points for specific countries: {xvalues}. The respective burnt area values are {yvalues}.",
    "Assemble a {chart_type} chart executing ChartJS titled '{title}' to present data as follows: {xvalue_yvalue_key_pairs}.",
    "Develop a {chart_type} chart utilizing ChartJS, titled '{title}'. The chart will present data for specific countries: {xvalues}, along with the corresponding burnt area values: {yvalues}.",
    "Build a {chart_type} chart utilizing ChartJS and name it '{title}'. This chart will highlight data points for the countries: {xvalues}, with the respective burnt area values being {yvalues}.",
    "Construct a {chart_type} chart utilizing ChartJS, named '{title}'. This chart will show data for the countries: {xvalues}, with the corresponding burnt area values being {yvalues}.",
    "Assemble a {chart_type} chart utilizing ChartJS and title it '{title}'. This visualization will highlight data points for specific countries: {xvalues}, with their respective burnt area values being {yvalues}.",
    "Develop a {chart_type} chart utilizing ChartJS, named '{title}'. The chart will display data points for significant countries: {xvalues}, with their corresponding burnt area values: {yvalues}.",
    "Formulate a {chart_type} chart utilizing ChartJS, titled '{title}'. This graph will present data points for given countries: {xvalues}, with their respective burnt area values being {yvalues}.",
    "Assemble a {chart_type} chart executing ChartJS titled '{title}' to present data as follows: {xvalue_yvalue_key_pairs}.",
    "Generate a {chart_type} chart utilizing ChartJS, titled '{title}'. This chart will illustrate data points for key countries: {xvalues}, and the burnt area values for these countries are {yvalues}.",
    "Design a {chart_type} graph utilizing ChartJS, named '{title}'. This chart will show data points for particular countries: {xvalues}, and their respective burnt area values are {yvalues}.",
    "Assemble a {chart_type} chart executing ChartJS titled '{title}' to present data as follows: {xvalue_yvalue_key_pairs}.",
    "Develop a {chart_type} chart utilizing ChartJS, titled '{title}'. The graph will represent data points for the countries: {xvalues}, with corresponding burnt area values being {yvalues}.",
    "Assemble a {chart_type} chart utilizing ChartJS, named '{title}'. This visualization will display data points for the following countries: {xvalues}, with the associated burnt area values being {yvalues}.",
    "Create a {chart_type} chart utilizing ChartJS, titled '{title}'. This chart will show data points for certain countries: {xvalues}, and the corresponding burnt area values are {yvalues}.",
    "Assemble a {chart_type} chart executing ChartJS titled '{title}' to present data as follows: {xvalue_yvalue_key_pairs}."
]

templates_bar_charts_nba = [
    "Create a bar chart using ChartJS. The chart should be titled '{title}'. It will display data points for players: {xvalues}. The corresponding statistics for these players are {yvalues} respectively.",
    "Craft a visual representation with ChartJS as a bar chart, ensuring the title reads '{title}'. This chart will showcase the top players and their respective statistics: {xvalue_yvalue_key_pairs}.",
    "Generate a bar chart using ChartJS. Name it '{title}' and display the performance data for the players: {xvalues}. The associated statistics are as follows: {yvalues}.",
    "Design a bar chart using ChartJS titled '{title}'. This chart will represent the performance metrics for specific players: {xvalues}. The corresponding values are {yvalues}.",
    "Produce a visual representation using ChartJS as a bar chart, with the title '{title}'. The chart will highlight player performance data: {xvalue_yvalue_key_pairs}.",
    "Construct a bar chart using ChartJS named '{title}'. This visualization will display the player statistics for: {xvalues}. The associated figures are {yvalues}, respectively.",
    "Create a bar chart using ChartJS titled '{title}'. It will showcase player performance markers: {xvalues}. The corresponding values are as follows: {yvalues}.",
    "Build a bar chart with ChartJS, titled '{title}'. This chart will show the statistics for players: {xvalues}. The associated values are {yvalues} respectively."
]
templates_pie_charts_nba = [
    "Create a pie chart using ChartJS. Title it '{title}'. The chart will show the percentage distribution among players: {xvalues}. Their respective values are {yvalues}.",
    "Design a visual representation with ChartJS as a pie chart titled '{title}'. This chart will display the share of each player in the dataset: {xvalue_yvalue_key_pairs}.",
    "Generate a pie chart with ChartJS named '{title}'. The chart will depict the percentage breakdown for players: {xvalues}. Their values are {yvalues}.",
    "Craft a pie chart using ChartJS, titled '{title}'. This visualization will illustrate the proportion of each player's statistics: {xvalues}. The corresponding values are {yvalues}.",
    "Create a pie chart using ChartJS with the title '{title}'. It will show the distribution of values among players: {xvalue_yvalue_key_pairs}.",
    "Construct a pie chart with ChartJS titled '{title}'. This chart will present the percentage of each player in the dataset: {xvalues}. Their values are {yvalues}.",
    "Build a pie chart using ChartJS named '{title}'. The chart will illustrate the distribution among players: {xvalues}. Their respective values are {yvalues}.",
    "Produce a pie chart using ChartJS with the title '{title}'. This chart will display the percentage share for players: {xvalues}. Their associated values are {yvalues}."
]

templates_bar_charts_movie = [
    "Create a bar chart using ChartJS. The chart should be titled '{title}'. It will display data points for movies: {xvalues}. The corresponding values are {yvalues} respectively.",
    "Craft a visual representation with ChartJS, ensuring the title reads '{title}'. This bar chart will showcase different movies and their respective values: {xvalue_yvalue_key_pairs}.",
    "Generate a bar chart using ChartJS. Name it '{title}' and display the data for the movies: {xvalues}. The associated values are as follows: {yvalues}.",
    "Design a bar chart using ChartJS titled '{title}'. This chart will represent the metrics for specific movies: {xvalues}. The corresponding values are {yvalues}.",
    "Produce a visual representation using ChartJS as a bar chart, with the title '{title}'. The chart will highlight movie data: {xvalue_yvalue_key_pairs}.",
    "Construct a bar chart using ChartJS named '{title}'. This visualization will display the statistics for movies: {xvalues}. The associated figures are {yvalues}, respectively.",
    "Create a bar chart using ChartJS titled '{title}'. It will showcase movie metrics: {xvalues}. The corresponding values are as follows: {yvalues}.",
    "Build a bar chart with ChartJS, titled '{title}'. This chart will show the data for movies: {xvalues}. The associated values are {yvalues} respectively."
]

templates_pie_charts_movie = [
    "Create a pie chart using ChartJS. Title it '{title}'. The chart will show the percentage distribution among movies: {xvalues}. Their respective values are {yvalues}.",
    "Design a visual representation with ChartJS titled '{title}'. This pie chart will display the share of each movie: {xvalue_yvalue_key_pairs}.",
    "Generate a pie chart with ChartJS named '{title}'. The chart will depict the percentage breakdown for movies: {xvalues}. Their values are {yvalues}.",
    "Craft a pie chart using ChartJS, titled '{title}'. This visualization will illustrate the proportion of each movie's metrics: {xvalues}. The corresponding values are {yvalues}.",
    "Create a pie chart using ChartJS with the title '{title}'. It will show the distribution of values among movies: {xvalue_yvalue_key_pairs}.",
    "Construct a pie chart with ChartJS titled '{title}'. This chart will present the percentage of each movie: {xvalues}. Their values are {yvalues}.",
    "Build a pie chart using ChartJS named '{title}'. The chart will illustrate the distribution among movies: {xvalues}. Their respective values are {yvalues}.",
    "Produce a pie chart using ChartJS with the title '{title}'. This chart will display the percentage share for movies: {xvalues}. Their associated values are {yvalues}."
]

templates_bar_charts_cars = [
    "Create a bar chart using ChartJS. The chart should be titled '{title}'. It will display data points for car models: {xvalues}. The corresponding values are {yvalues} respectively.",
    "Craft a visual representation with ChartJS as a bar chart, ensuring the title reads '{title}'. This chart will showcase different car models and their respective values: {xvalue_yvalue_key_pairs}.",
    "Generate a bar chart using ChartJS. Name it '{title}' and display the data for the car models: {xvalues}. The associated values are as follows: {yvalues}.",
    "Design a bar chart using ChartJS titled '{title}'. This chart will represent the metrics for specific car models: {xvalues}. The corresponding values are {yvalues}.",
    "Produce a visual representation using ChartJS, with the title '{title}'. The chart will highlight car model data: {xvalue_yvalue_key_pairs}.",
    "Construct a bar chart using ChartJS named '{title}'. This visualization will display the statistics for car models: {xvalues}. The associated figures are {yvalues}, respectively.",
    "Create a bar chart using ChartJS titled '{title}'. It will showcase car model metrics: {xvalues}. The corresponding values are as follows: {yvalues}.",
    "Build a bar chart with ChartJS, titled '{title}'. This chart will show the data for car models: {xvalues}. The associated values are {yvalues} respectively."
]

templates_pie_charts_cars = [
    "Create a pie chart using ChartJS. Title it '{title}'. The chart will show the percentage distribution among car models: {xvalues}. Their respective values are {yvalues}.",
    "Design a visual representation with ChartJS as a pie chart titled '{title}'. This chart will display the share of each car model: {xvalue_yvalue_key_pairs}.",
    "Generate a pie chart with ChartJS named '{title}'. The chart will depict the percentage breakdown for car models: {xvalues}. Their values are {yvalues}.",
    "Craft a pie chart using ChartJS, titled '{title}'. This visualization will illustrate the proportion of each car model's metrics: {xvalues}. The corresponding values are {yvalues}.",
    "Create a pie chart using ChartJS with the title '{title}'. It will show the distribution of values among car models: {xvalue_yvalue_key_pairs}.",
    "Construct a pie chart with ChartJS titled '{title}'. This chart will present the percentage of each car model: {xvalues}. Their values are {yvalues}.",
    "Build a pie chart using ChartJS named '{title}'. The chart will illustrate the distribution among car models: {xvalues}. Their respective values are {yvalues}.",
    "Produce a pie chart using ChartJS with the title '{title}'. This chart will display the percentage share for car models: {xvalues}. Their associated values are {yvalues}."
]

templates_bar_charts_blood = [
    "Create a bar chart using ChartJS. The chart should be titled '{title}'. It will display data points for blood types: {xvalues}. The corresponding values are {yvalues} respectively.",
    "Craft a visual representation with ChartJS as a bar chart, ensuring the title reads '{title}'. This chart will showcase different blood types and their respective quantities: {xvalue_yvalue_key_pairs}.",
    "Generate a bar chart using ChartJS. Name it '{title}' and display the data for the blood types: {xvalues}. The associated quantities are as follows: {yvalues}.",
    "Design a bar chart using ChartJS titled '{title}'. This chart will represent the quantities for specific blood types: {xvalues}. The corresponding values are {yvalues}.",
    "Produce a visual representation using ChartJS, with the title '{title}'. The chart will highlight blood type data: {xvalue_yvalue_key_pairs}.",
    "Construct a bar chart using ChartJS named '{title}'. This visualization will display the quantities for blood types: {xvalues}. The associated figures are {yvalues}, respectively.",
    "Create a bar chart using ChartJS titled '{title}'. It will showcase blood type quantities: {xvalues}. The corresponding values are as follows: {yvalues}.",
    "Build a bar chart with ChartJS, titled '{title}'. This chart will show the data for blood types: {xvalues}. The associated quantities are {yvalues} respectively."
]

templates_pie_charts_blood = [
    "Create a pie chart using ChartJS. Title it '{title}'. The chart will show the distribution among blood types: {xvalues}. Their respective quantities are {yvalues}.",
    "Design a visual representation with ChartJS as a pie chart titled '{title}'. This chart will display the share of each blood type: {xvalue_yvalue_key_pairs}.",
    "Generate a pie chart with ChartJS named '{title}'. The chart will depict the breakdown for blood types: {xvalues}. Their quantities are {yvalues}.",
    "Craft a pie chart using ChartJS, titled '{title}'. This visualization will illustrate the proportion of each blood type's quantities: {xvalues}. The corresponding values are {yvalues}.",
    "Create a pie chart using ChartJS with the title '{title}'. It will show the distribution of quantities among blood types: {xvalue_yvalue_key_pairs}.",
    "Construct a pie chart with ChartJS titled '{title}'. This chart will present the quantities of each blood type: {xvalues}. Their values are {yvalues}.",
    "Build a pie chart using ChartJS named '{title}'. The chart will illustrate the distribution among blood types: {xvalues}. Their respective values are {yvalues}.",
    "Produce a pie chart using ChartJS with the title '{title}'. This chart will display the quantitative share for blood types: {xvalues}. Their associated quantities are {yvalues}."
]

templates_bar_charts_demographics = [
    "Create a bar chart using ChartJS. The chart should be titled '{title}'. It will display data points for demographics: {xvalues}. The corresponding values are {yvalues} respectively.",
    "Craft a visual representation with ChartJS as a bar chart, ensuring the title reads '{title}'. This chart will showcase different demographic categories and their respective counts: {xvalue_yvalue_key_pairs}.",
    "Generate a bar chart using ChartJS. Name it '{title}' and display the data for the demographics: {xvalues}. The associated figures are as follows: {yvalues}.",
    "Design a bar chart using ChartJS titled '{title}'. This chart will represent the totals for specific demographics: {xvalues}. The corresponding values are {yvalues}.",
    "Produce a visual representation using ChartJS, with the title '{title}'. The chart will highlight demographic data: {xvalue_yvalue_key_pairs}.",
    "Construct a bar chart using ChartJS named '{title}'. This visualization will display the numbers for demographics: {xvalues}. The associated figures are {yvalues}, respectively.",
    "Create a bar chart using ChartJS titled '{title}'. It will showcase demographic counts: {xvalues}. The corresponding values are as follows: {yvalues}.",
    "Build a bar chart with ChartJS, titled '{title}'. This chart will show the data for demographics: {xvalues}. The associated numbers are {yvalues} respectively."
]

templates_pie_charts_demographics = [
    "Create a pie chart using ChartJS. Title it '{title}'. The chart will show the distribution among demographics: {xvalues}. Their respective figures are {yvalues}.",
    "Design a visual representation with ChartJS as a pie chart titled '{title}'. This chart will display the share of each demographic category: {xvalue_yvalue_key_pairs}.",
    "Generate a pie chart with ChartJS named '{title}'. The chart will depict the breakdown for demographics: {xvalues}. Their figures are {yvalues}.",
    "Craft a pie chart using ChartJS, titled '{title}'. This visualization will illustrate the proportion of each demographic category's totals: {xvalues}. The corresponding values are {yvalues}.",
    "Create a pie chart using ChartJS with the title '{title}'. It will show the distribution of totals among demographics: {xvalue_yvalue_key_pairs}.",
    "Construct a pie chart with ChartJS titled '{title}'. This chart will present the number of each demographic category: {xvalues}. Their figures are {yvalues}.",
    "Build a pie chart using ChartJS named '{title}'. The chart will illustrate the distribution among demographics: {xvalues}. Their respective counts are {yvalues}.",
    "Produce a pie chart using ChartJS with the title '{title}'. This chart will display the share for demographics: {xvalues}. Their associated figures are {yvalues}."
]

templates_bar_charts_leetcode = [
    "Create a bar chart using ChartJS. The chart should be titled '{title}'. It will display data points for LeetCode problems: {xvalues}. The corresponding numbers are {yvalues} respectively.",
    "Craft a visual representation with ChartJS as a bar chart, ensuring the title reads '{title}'. This chart will showcase different types of LeetCode problems and their respective counts: {xvalue_yvalue_key_pairs}.",
    "Generate a bar chart using ChartJS. Name it '{title}' and display the data for the types of LeetCode problems: {xvalues}. The associated figures are as follows: {yvalues}.",
    "Design a bar chart using ChartJS titled '{title}'. This chart will represent the totals for specific types of LeetCode problems: {xvalues}. The corresponding numbers are {yvalues}.",
    "Produce a visual representation using ChartJS, with the title '{title}'. The chart will highlight data on LeetCode problems: {xvalue_yvalue_key_pairs}.",
    "Construct a bar chart using ChartJS named '{title}'. This visualization will display the numbers for types of LeetCode problems: {xvalues}. The associated figures are {yvalues}, respectively.",
    "Create a bar chart using ChartJS titled '{title}'. It will showcase counts of different types of LeetCode problems: {xvalues}. The corresponding numbers are as follows: {yvalues}.",
    "Build a bar chart with ChartJS, titled '{title}'. This chart will show the data for types of LeetCode problems: {xvalues}. The associated numbers are {yvalues} respectively."
]

templates_pie_charts_leetcode = [
    "Create a pie chart using ChartJS. Title it '{title}'. The chart will show the distribution of problem types on LeetCode: {xvalues}. Their corresponding numbers are {yvalues}.",
    "Design a visual representation with ChartJS as a pie chart titled '{title}'. This chart will display the share of each type of problem on LeetCode: {xvalue_yvalue_key_pairs}.",
    "Generate a pie chart with ChartJS named '{title}'. The chart will depict the breakdown of problem types on LeetCode: {xvalues}. Their numbers are {yvalues}.",
    "Craft a pie chart using ChartJS, titled '{title}'. This visualization will illustrate the proportion of each type of problem's numbers on LeetCode: {xvalues}. The corresponding values are {yvalues}.",
    "Create a pie chart using ChartJS with the title '{title}'. It will show the distribution of numbers among problem types on LeetCode: {xvalue_yvalue_key_pairs}.",
    "Construct a pie chart with ChartJS titled '{title}'. This chart will present the number of each type of problem on LeetCode: {xvalues}. Their numbers are {yvalues}.",
    "Build a pie chart using ChartJS named '{title}'. The chart will illustrate the distribution of problem types on LeetCode: {xvalues}. Their corresponding counts are {yvalues}.",
    "Produce a pie chart using ChartJS with the title '{title}'. This chart will display the share of problem types on LeetCode: {xvalues}. Their associated numbers are {yvalues}."
]

templates_bar_charts_population = [
    "Create a bar chart using ChartJS. The chart should be titled '{title}'. It will display data points for country populations: {xvalues}. The corresponding populations are {yvalues} respectively.",
    "Craft a visual representation with ChartJS as a bar chart, ensuring the title reads '{title}'. This chart will showcase populations of different countries: {xvalue_yvalue_key_pairs}.",
    "Generate a bar chart using ChartJS. Name it '{title}' and display the data for country populations: {xvalues}. The associated numbers are as follows: {yvalues}.",
    "Design a bar chart using ChartJS titled '{title}'. This chart will represent the population totals for specific countries: {xvalues}. The corresponding populations are {yvalues}.",
    "Produce a visual representation using ChartJS, with the title '{title}'. The chart will highlight country population data: {xvalue_yvalue_key_pairs}.",
    "Construct a bar chart using ChartJS named '{title}'. This visualization will display the population numbers for countries: {xvalues}. The associated populations are {yvalues}, respectively.",
    "Create a bar chart using ChartJS titled '{title}'. It will showcase population counts for different countries: {xvalues}. The corresponding populations are as follows: {yvalues}.",
    "Build a bar chart with ChartJS, titled '{title}'. This chart will show the data for country populations: {xvalues}. The associated populations are {yvalues} respectively."
]

templates_pie_charts_population = [
    "Create a pie chart using ChartJS. Title it '{title}'. The chart will show the distribution of country populations: {xvalues}. Their corresponding populations are {yvalues}.",
    "Design a visual representation with ChartJS as a pie chart titled '{title}'. This chart will display the share of population for each country: {xvalue_yvalue_key_pairs}.",
    "Generate a pie chart with ChartJS named '{title}'. The chart will depict the breakdown of populations across countries: {xvalues}. Their populations are {yvalues}.",
    "Craft a pie chart using ChartJS, titled '{title}'. This visualization will illustrate the proportion of each country's population: {xvalues}. The corresponding populations are {yvalues}.",
    "Create a pie chart using ChartJS with the title '{title}'. It will show the distribution of populations among countries: {xvalue_yvalue_key_pairs}.",
    "Construct a pie chart with ChartJS titled '{title}'. This chart will present the population numbers for each country: {xvalues}. Their populations are {yvalues}.",
    "Build a pie chart using ChartJS named '{title}'. The chart will illustrate the distribution of country populations: {xvalues}. Their corresponding counts are {yvalues}.",
    "Produce a pie chart using ChartJS with the title '{title}'. This chart will display the share of populations among countries: {xvalues}. Their associated populations are {yvalues}."
]

templates_bar_charts_cancer = [
    "Create a bar chart using ChartJS. The chart should be titled '{title}'. It will display data points for cancer statistics: {xvalues}. The corresponding numbers are {yvalues} respectively.",
    "Craft a visual representation with ChartJS as a bar chart, ensuring the title reads '{title}'. This chart will showcase different types of cancer and their respective cases: {xvalue_yvalue_key_pairs}.",
    "Generate a bar chart using ChartJS. Name it '{title}' and display the data for cancer statistics: {xvalues}. The associated figures are as follows: {yvalues}.",
    "Design a bar chart using ChartJS titled '{title}'. This chart will represent the total cases for specific types of cancer: {xvalues}. The corresponding numbers are {yvalues}.",
    "Produce a visual representation using ChartJS, with the title '{title}'. The chart will highlight cancer statistics: {xvalue_yvalue_key_pairs}.",
    "Construct a bar chart using ChartJS named '{title}'. This visualization will display the numbers for types of cancer: {xvalues}. The associated figures are {yvalues}, respectively.",
    "Create a bar chart using ChartJS titled '{title}'. It will showcase case counts for different types of cancer: {xvalues}. The corresponding numbers are as follows: {yvalues}.",
    "Build a bar chart with ChartJS, titled '{title}'. This chart will show the data for cancer statistics: {xvalues}. The associated numbers are {yvalues} respectively."
]

templates_pie_charts_cancer = [
    "Create a pie chart using ChartJS. Title it '{title}'. The chart will show the distribution of cancer cases: {xvalues}. Their corresponding numbers are {yvalues}.",
    "Design a visual representation with ChartJS as a pie chart titled '{title}'. This chart will display the share of cases for each type of cancer: {xvalue_yvalue_key_pairs}.",
    "Generate a pie chart with ChartJS named '{title}'. The chart will depict the breakdown of cases across types of cancer: {xvalues}. Their numbers are {yvalues}.",
    "Craft a pie chart using ChartJS, titled '{title}'. This visualization will illustrate the proportion of each type of cancer's cases: {xvalues}. The corresponding numbers are {yvalues}.",
    "Create a pie chart using ChartJS with the title '{title}'. It will show the distribution of cases among types of cancer: {xvalue_yvalue_key_pairs}.",
    "Construct a pie chart with ChartJS titled '{title}'. This chart will present the case numbers for each type of cancer: {xvalues}. Their numbers are {yvalues}.",
    "Build a pie chart using ChartJS named '{title}'. The chart will illustrate the distribution of cancer cases: {xvalues}. Their corresponding counts are {yvalues}.",
    "Produce a pie chart using ChartJS with the title '{title}'. This chart will display the share of cases among types of cancer: {xvalues}. Their associated numbers are {yvalues}."
]



# function that chooses a random prompt template
def pick_a_random_template(template_list):
    return random.choice(template_list)

# function to know what prompt to use
def get_template_type(json_data_point):
    normalized_title = json_data_point['chart_title'].lower()
    if 'blood' in normalized_title:
        if json_data_point['chart_type'] == 'pie':
            return templates_pie_charts_blood
        return templates_bar_charts_blood
    elif 'male' in normalized_title or 'female' in normalized_title:
        if json_data_point['chart_type'] == 'pie':
            return templates_pie_charts_demographics
        return templates_bar_charts_demographics
    elif 'leetcode' in normalized_title:
        if json_data_point['chart_type'] == 'pie':
            return templates_pie_charts_leetcode
        return templates_bar_charts_leetcode
    elif 'population' in normalized_title:
        if json_data_point['chart_type'] == 'pie':
            return templates_pie_charts_population
        return templates_bar_charts_population
    elif 'nba' in normalized_title:
        if json_data_point['chart_type'] == 'pie':
            return templates_pie_charts_nba
        return templates_bar_charts_nba
    elif 'cars' in normalized_title:
        if json_data_point['chart_type'] == 'pie':
            return templates_pie_charts_cars
        return templates_bar_charts_cars
    elif 'movie' in normalized_title:
        if json_data_point['chart_type'] == 'pie':
            return templates_pie_charts_movie
        return templates_bar_charts_movie
    elif 'stock' in normalized_title:
        return templates_line_charts
    else:
        if json_data_point['chart_type'] == 'pie':
            return templates_pie_charts_cancer
        return templates_bar_charts_cancer
    

# function to insert an extracted chart type into the template 
def insert_chart_type(target_string, chart_type):
    res = target_string.replace("{chart_type}", chart_type)
    return res
    
# function to insert the extracted chart title into the template   
def insert_chart_title(target_string, chart_title):
    res = target_string.replace("{title}", chart_title)
    return res

# crutch works for the first iteration of my line charts(not universal)
# function to insert date format labels into the template
def insert_dates(target_string, dates):
    # random choice between 0 and 1 signifies which date format will be chosen
    date_format = random.choice([0, 1])
    # converting string dates into datetime object
    dates_in_datetime = [datetime.strptime(date, "%Y-%m-%d") for date in dates]
    dates_to_export = []
    # if random choice is 0 then format will me in the YYYY-MM-DD format
    if(date_format == 0):
        dates_to_export = [datetime.strftime(date, "%Y-%m-%d") for date in dates_in_datetime]
    # if it is equal to 1 then Month Day, Year format will be chosen
    else:
        dates_to_export = [datetime.strftime(date, "%B %d, %Y") for date in dates_in_datetime]
    # then through a loop the datetime objects get converted to string
    str_dts = ""
    for date in dates_to_export:
        str_dts = str_dts + date + ", " 
    # getting rid of not needed spaces at the end
    str_dts = str_dts.strip()
    # and grabbing every character instead of the last comma
    str_dts = str_dts[:-1]
    # formats = []
    res = target_string.replace("{dates}", str_dts)
    return res

# function to insert any kind of string label into the template
def insert_string_xvalues(target_string, xvalues):
    # same procedure as before
    temp = ""
    for xval in xvalues:
        temp = temp + xval + ", "
    temp = temp.strip()
    temp = temp[:-1]
    res = target_string.replace("{xvalues}", temp)
    return res

# crutch works for the first iteration of my line charts(not universal)
# functions insert_values and insert_yvalues are almost identical(they insert yvalue data for the plot) however
# the replace values are slightly different
# why this is the case because the insert_values function is based on the 
# old standard and It works and I do not want to change it
def insert_values(target_string, values):
    str_vls = ""
    for val in values:
        str_vls = str_vls + str(val) + ", " 
    str_vls = str_vls.strip()
    str_vls = str_vls[:-1]
    res = target_string.replace("{values}", str_vls)
    return res
    
def insert_yvalues(target_string, yvalues):
    str_vls = ""
    for val in yvalues:
        str_vls = str_vls + str(val) + ", " 
    str_vls = str_vls.strip()
    str_vls = str_vls[:-1]
    res = target_string.replace("{yvalues}", str_vls)
    return res

# some of the prompts have a different format and they are based on a key value concept
# and function is used to insert those values
def insert_date_value_pairs(target_string, dates, values):
    date_format = random.choice([0, 1])
    dates_in_datetime = [datetime.strptime(date, "%Y-%m-%d") for date in dates]
    dates_to_export = []
    #example of date format swap
    if(date_format == 0):
        dates_to_export = [datetime.strftime(date, "%Y-%m-%d") for date in dates_in_datetime]
    else:
        dates_to_export = [datetime.strftime(date, "%B %d, %Y") for date in dates_in_datetime]
    # connecting_words list contains linking words between the date and the value
    connecting_words = ["at", "with a value of", "equaling to", "being at"]
    kvpair = ""
    for i in range(0, len(dates_to_export)):
        kvpair = kvpair + dates_to_export[i] + " " + random.choice(connecting_words) + " " + str(values[i]) + "; "
    kvpair = kvpair.strip()
    res = target_string.replace("{date_value_key_pairs}", kvpair)
    return res

# inserting key value for any string
def insert_x_y_value_pairs(target_string, xvalues, yvalues):
    connecting_words = ["at", "with a value of", "equaling to", "being at"]
    kvpair = ""
    for i in range(0, len(xvalues)):
        kvpair = kvpair + xvalues[i] + " " + random.choice(connecting_words) + " " + str(yvalues[i]) + "; "
    kvpair = kvpair.strip()
    res = target_string.replace("{xvalue_yvalue_key_pairs}", kvpair)
    return res

# calling all insertion functions in a neatly manner
def insert_all(target_string, chart_type, chart_title, dates, values):
    res = insert_chart_type(target_string, chart_type)
    res = insert_chart_title(res, chart_title)
    if "{date_value_key_pairs}" in target_string:
        res = insert_date_value_pairs(res, dates, values)
    else:
        res = insert_dates(res, dates)
        res = insert_values(res, values)
    return res

# calling all universal insertion functions
def insert_all_universal(target_string, chart_type, chart_title, xvalues, yvalues):
    res = insert_chart_type(target_string, chart_type)
    res = insert_chart_title(res, chart_title)
    if "{xvalue_yvalue_key_pairs}" in target_string:
        res = insert_x_y_value_pairs(res, xvalues, yvalues)
    else:
        res = insert_string_xvalues(res, xvalues)
        res = insert_yvalues(res, yvalues)
    return res

# swapping non essential data for prompt variety with synonyms
def synonym_generator(target_string):
    prompt = target_string
    # common words that can be changed
    synonyms = {
    "Generate": ["Create", "Produce", "Construct", "Craft", "Make", "Formulate", "Design", "Assemble", "Build", "Generate"],
    "utilizing": ["employing", "leveraging", "implementing", "incorporating", "using", "utilizing"],
    "chart": ["graph", "diagram", "plot", "chart"],
    "titled": ["named", "called", "identified as", "titled"],
    "calling it": ["naming it", "identifying it as", "calling it"],
    "will": ["is going to", "will"],
    "display": ["showcase", "represent", "present", "illustrate", "display"],
    }
    # iterate through the dictionary
    for word, syn_list in synonyms.items():
        # if the dictionary key is present in the prompt a random synonym is chosen
        # from its values and inserted into the prompt
        if word in prompt:
            synonym = random.choice(syn_list)
            prompt = prompt.replace(word, synonym)
        # if not then synonym lists are scanned to find the word that is present in the prompt
        # which is then swapped
        else:
            #print(syn_list)
            for syn in syn_list:
                if syn in prompt:
                    word = syn
                    synonym = random.choice(syn_list)
                    prompt = prompt.replace(word, synonym)
                    break
    return prompt
```

```python
def validate_date_answers(answer, dates, values):
    # Define regular expressions for dates and floating point numbers
    date_pattern = r'\b(?:\d{4}-\d{2}-\d{2}|\b(?:January|February|March|April|May|June|July|August|September|October|November|December) \d{1,2}, \d{4}|\b\d{1,2}/\d{1,2}/\d{4})\b'
    float_pattern = r'\d+\.\d+'
    
    # Extract dates
    extracted_dates = re.findall(date_pattern, answer)
    
    # Extract floating point numbers
    extracted_values = re.findall(float_pattern, answer)
    
    # Convert values to float
    values = list(map(float, values))

    dates_mdyy = [datetime.strptime(extracted_date, "%B %d, %Y") for extracted_date in extracted_dates]
    dates_yyyymmdd = [datetime.strptime(original_date, "%Y-%m-%d") for original_date in dates]
    comparison_results = []
    for i in range(0, len(dates_mdyy)):
        if (dates_mdyy[i] == dates_yyyymmdd[i]) and (extracted_values[i] == values[i]):
            comparison_results.append((dates_mdyy[i], "Match found"))
        else:
            comparison_results.append((dates_mdyy[i], "No match"))
            
    
    # Print comparison results
    #print("\nComparison results:")
    for result in comparison_results:
        print(f"Date: {result[0].strftime('%B %d, %Y')} - {result[1]}")

    print("Dates:", dates)
    print("Values:", values)
# for prompt in prompts:
#     validate_date_answers(output[0]['generated_text'], labels_list_names, labels_list_values)
```
