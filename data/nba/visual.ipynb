```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = './output.csv'
df = pd.read_csv(file_path)
```

```python
# Visualization of column1 as a bar chart
plt.figure(figsize=(10, 6))
df.head()
df['Chart Type'].value_counts().plot(kind='bar')
plt.title('Bar Chart of Chart')
plt.xlabel('Chart Type')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization of column2 as a bar chart
plt.figure(figsize=(10, 6))
df['Data Points'] = df['Data Points'].astype(int)
sort_index = df['Data Points'].value_counts().sort_index()
sort_index.plot(kind='bar')
plt.title('Bar Chart of DP')
plt.xlabel('Number of Labels')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
unique_chart_types = df['Chart Type'].unique()

for chart_type in unique_chart_types:
    plt.figure(figsize=(10, 6))
    subset = df[df['Chart Type'] == chart_type]
    sort_index = subset['Data Points'].value_counts().sort_index()
    sort_index.plot(kind='bar')
    plt.title(f'Bar Chart of Data Points for {chart_type}')
    plt.xlabel('Number of Labels')
    plt.ylabel('Count')
    plt.show()
```
