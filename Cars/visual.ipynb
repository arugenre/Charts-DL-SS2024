```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = './output_cars_charts.csv'
df = pd.read_csv(file_path)
```

```python
# Visualization of column1 as a bar chart
plt.figure(figsize=(10, 6))
df.head()
df['Chart Type'].value_counts().plot(kind='bar')
plt.title('Bar Chart of Chart')
plt.xlabel('chart')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization of column2 as a bar chart
plt.figure(figsize=(10, 6))
df['Data Points'] = df['Data Points'].astype(int)
sort_index = df['Data Points'].value_counts().sort_index()
sort_index.plot(kind='bar')
plt.xlabel('column2')
plt.ylabel('Count')
plt.show()
```
